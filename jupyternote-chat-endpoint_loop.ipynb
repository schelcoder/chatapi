```python
# Import openai
from openai import OpenAI


summaries = []

for row in df.itertuples():
    # Access attributes using row.column_name
    print(row.company)

    content = "can you summarize in less then 100 words does the following company do?" + row.company
 

    client = OpenAI(
        # defaults to os.environ.get("OPENAI_API_KEY")
        api_key="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",
    )

    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": content,
            }
        ],
        model="gpt-4-1106-preview",
    )

    summaries.append(chat_completion)





```

```python
print(summaries[100].choices[0].message.content)

type(summaries)
```

```python
df = pd.DataFrame(columns=['name','summary'])

companies = pd.read_csv('../Environment_sustainability_businesses.csv', delimiter='\r')

for i in range(len(summaries)):
#for i in range(0,6):

    new_record = pd.DataFrame([{'name':companies.iloc[i,0], 'summary':summaries[i].choices[0].message.content}])
    df = pd.concat([df, new_record], ignore_index=True)

print(df)
```
